```python
import cv2
import numpy as np
from skimage import exposure
from matplotlib import pyplot as plt
```

```python
image_source = cv2.imread('img.png', cv2.IMREAD_GRAYSCALE)
image_target = cv2.imread('img_1.png', cv2.IMREAD_GRAYSCALE)
```

```python
matched_image = exposure.match_histograms(image_source, image_target)
plt.figure(figsize=(10, 5))
```

```python
plt.subplot(3, 3, 1)
plt.imshow(image_source, cmap='gray')
plt.title('Source Image')
plt.axis('off')
```

```python
plt.subplot(3, 3, 2)
plt.hist(image_target.ravel(), bins=256)
plt.title('Target Image')
plt.axis('off')
```

```python
plt.subplot(3, 3, 4)
plt.imshow(image_target, cmap='gray')
plt.title('Target Image')
```

```python
plt.subplot(3,3,5)
plt.hist(image_target.ravel(), bins=256)
plt.title('Matched Image')
plt.axis('off')
```

```python
plt.subplot(3,3,7)
plt.imshow(matched_image, cmap='gray')
plt.title('Matched Image')
plt.axis('off')
```
