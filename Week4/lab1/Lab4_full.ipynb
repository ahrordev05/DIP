```python
import cv2, numpy as np, matplotlib.pyplot as plt, os
```

```python

def gamma_correction(img, c, gamma):
    I_norm = img.astype('float32')/255
    out = c*(I_norm**gamma)
    return np.clip(out*255,0,255).astype('uint8')
```

```python

def contrast_stretching(img, smin, smax):
    rmin, rmax = img.min(), img.max()
    s = (smax-smin)/(rmax-rmin)*(img-rmin)+smin
    return np.clip(s,0,255).astype('uint8')
```

```python

folder="lab04_assets/images"
os.makedirs("outputs",exist_ok=True)

for fname in os.listdir(folder):
    img=cv2.imread(os.path.join(folder,fname),cv2.IMREAD_GRAYSCALE)
    g1=gamma_correction(img,1,0.5)
    g2=gamma_correction(img,1,2)
    c1=contrast_stretching(img,0,255)
    c2=contrast_stretching(img,220,50)
    c3=contrast_stretching(img,10,420)
    c4=contrast_stretching(img,90,555)
    cv2.imwrite(f"outputs/gamma05_{fname}",g1)
    cv2.imwrite(f"outputs/gamma2_{fname}",g2)
    cv2.imwrite(f"outputs/contrast_{fname}",c1)
    cv2.imwrite(f"outputs/contrast50_200_{fname}",c2)
    cv2.imwrite(f"outputs/contrast10_420_{fname}",c3)
    cv2.imwrite(f"outputs/contrast90_555_{fname}",c4)

print("All images processed and saved in outputs/")
```
